# Exploratory Statistics with Python (Everyday Example: Peopleâ€™s Ages)

```python
# ðŸ“Š Day 1: Exploratory Statistics with Python
# Author: David Caleb Chaparro Orozco
# Topic: Everyday example - Age analysis

# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Set default style for plots
sns.set(style="whitegrid")
```

In everyday life, we constantly make decisions based on data â€” whether comparing salaries, tracking personal expenses, or evaluating health indicators like weight or age.

In this notebook, we'll analyze a small dataset of **ages of people in a neighborhood** to learn how to extract **meaningful statistical insights** using Python.


```python
# Example: Ages of 20 people from a local community
ages = [25, 34, 30, 28, 22, 21, 40, 38, 60, 65, 70, 45, 33, 29, 27, 26, 32, 35, 30, 28]
data = pd.DataFrame({'Age': ages})
data.head()
```

### ðŸ“Œ Central tendency tells us where most values lie in a dataset.
Let's compute the:
- **Mean**: Average age
- **Median**: Middle value
- **Mode**: Most frequent age


```python
mean_age = data['Age'].mean()
median_age = data['Age'].median()
mode_age = data['Age'].mode().values[0]

print(f"Mean age: {mean_age:.2f} years")
print(f"Median age: {median_age} years")
print(f"Mode age: {mode_age} years")
```

### ðŸ“Œ Dispersion tells us how spread out the data is.
We'll look at:
- **Range**: Difference between oldest and youngest person
- **Standard Deviation**: Average distance from the mean
- **Variance**: Square of standard deviation

```python
range_age = data['Age'].max() - data['Age'].min()
std_age = data['Age'].std()
var_age = data['Age'].var()

print(f"Range of ages: {range_age} years")
print(f"Standard deviation: {std_age:.2f}")
print(f"Variance: {var_age:.2f}")
```

### ðŸ§  Visualizing data helps us understand its shape, skew, and outliers.
We'll use:
- **Histogram** to see frequency
- **Boxplot** to check outliers and quartiles


```python
plt.figure(figsize=(12, 5))

# Histogram
plt.subplot(1, 2, 1)
sns.histplot(data['Age'], bins=8, kde=True, color='skyblue')
plt.title('Histogram of Ages')

# Boxplot
plt.subplot(1, 2, 2)
sns.boxplot(x=data['Age'], color='lightgreen')
plt.title('Boxplot of Ages')

plt.tight_layout()
plt.savefig("outputs/histograms.png")
plt.show()
```

### ðŸ¤“ What do these terms mean?

- **Skewness**: Measures asymmetry (left or right tilt)
- **Kurtosis**: Measures whether data is heavy-tailed or light-tailed


```python
skewness = stats.skew(data['Age'])
kurtosis = stats.kurtosis(data['Age'])

print(f"Skewness: {skewness:.2f}")
print(f"Kurtosis: {kurtosis:.2f}")
```

```python
summary = data.describe()
summary.to_string("outputs/summary_statistics.txt")
summary
```

- The **mean age** is around 35.90 years, indicating the average adult age.
- The **median** helps when there are outliers (like someone aged 70).
- The **boxplot** shows the middle 50% and any extreme values.
- Skewness and kurtosis help us understand if the data is symmetric or peaked.
